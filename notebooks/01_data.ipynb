```python
import pandas as pd
import os
```

```python
DATA_PATH = 'C:/Users/angel/OneDrive/Desktop/ProgettoMeDL'
NEWS_FILENAME = 'C:/Users/angel/OneDrive/Desktop/ProgettoMeDL/nasdaq_exteral_data.csv'  
PRICE_FILENAME = 'C:/Users/angel/OneDrive/Desktop/ProgettoMeDL/full_history/full_history/VOO.csv' 
```

```python

df_tickers = pd.read_csv(os.path.join(DATA_PATH, NEWS_FILENAME), usecols=['Stock_symbol'])
unique_tickers = df_tickers['Stock_symbol'].unique()

target_indices = ['SPY']

print(f"Totale Ticker unici nel dataset: {len(unique_tickers)}")
print("-" * 30)

found_any = False
for target in target_indices:
    if target in unique_tickers:
        found_any = True
        count = df_tickers[df_tickers['Stock_symbol'] == target].shape[0]
        print(f" Numero di news: {count}")
    else:
        print(f"NON TROVATO: {target}")
```

```python
OUTPUT_FILE = 'C:/Users/angel/OneDrive/Desktop/ProgettoMeDL/SPY_news_only.csv'
TARGET_TICKER = 'SPY'
CHUNK_SIZE = 500000
spy_news_list = []

chunk_iter = pd.read_csv(NEWS_FILENAME, chunksize=CHUNK_SIZE)

count_processed = 0
for i, chunk in enumerate(chunk_iter):
    filtered_chunk = chunk[chunk['Stock_symbol'] == TARGET_TICKER]
    if not filtered_chunk.empty:
        spy_news_list.append(filtered_chunk)
    
    count_processed += len(chunk)
    if i % 5 == 0:
        print(f"Righe analizzate: {count_processed}...")
if spy_news_list:
    df_spy_news = pd.concat(spy_news_list, ignore_index=True)
    final_count = len(df_spy_news)
    print(f"\nEstrazione completata!")
    print(f"Notizie totali trovate per {TARGET_TICKER}: {final_count}")
    df_spy_news.to_csv(OUTPUT_FILE, index=False)
    print(f"File salvato con successo: {OUTPUT_FILE}")
else:
    print(f"\nAttenzione: Nessuna notizia trovata per {TARGET_TICKER}.")
```
